```python
# ====================================
# 1️⃣ DB 초기 설정
# ====================================
conn = sqlite3.connect("safetalk.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT UNIQUE,
    exp INTEGER DEFAULT 0,
    level INTEGER DEFAULT 1,
    total_score INTEGER DEFAULT 0,
    created_at TIMESTAMP DEFAULT CURRENT_TIMESTAMP
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS results (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    user_id INTEGER,
    scenario TEXT,
    score INTEGER,
    total INTEGER,
    date TIMESTAMP DEFAULT CURRENT_TIMESTAMP,
    FOREIGN KEY (user_id) REFERENCES users(id)
)
""")
conn.commit()

# ====================================
# 2️⃣ 시나리오 문제
# ====================================
scenarios = {
    "택배 스미싱": {
        "설명": "택배사·운송사를 사칭하여 클릭 유도나 개인 정보 입력을 요구하는 스미싱 유형입니다.",
        "문제": [
            {
                "질문": "📦 'CJ대한통운입니다. 배송지 오류로 반송 중입니다. 주소를 재입력해주세요.' 이런 문자를 받았다면?",
                "보기": {
                    1: "링크를 눌러 주소를 수정한다.",
                    2: "공식 고객센터(1588-1255)에 직접 문의한다.",
                    3: "문자를 캡처해 가족 단톡방에 공유한다."
                },
                "정답": 2
            },
            {
                "질문": "🚚 '국제특송 통관번호 오류로 반송 예정입니다'라는 문자는 어떤 심리를 노린 것일까요?",
                "보기": {
                    1: "호기심과 불안감",
                    2: "보상 욕구",
                    3: "공포감과 위협"
                },
                "정답": 1
            },
            {
                "질문": "📱 스미싱 문자에 자주 포함되는 단어가 아닌 것은?",
                "보기": {
                    1: "환불",
                    2: "이체",
                    3: "축하합니다"
                },
                "정답": 3
            },
        ]
    },

    "공공기관 사칭": {
        "설명": "경찰청, 국세청, 건강보험공단 등 정부기관을 사칭한 스미싱 유형입니다.",
        "문제": [
            {
                "질문": "🚨 '[경찰청교통민원] 교통범칙금 통지서 발송 완료'라는 문자를 받았다면?",
                "보기": {
                    1: "URL을 눌러 위반 사실을 확인한다.",
                    2: "문자를 무시하고 교통민원24 공식사이트에서 직접 확인한다.",
                    3: "문자를 캡처해서 지인에게 확인받는다."
                },
                "정답": 2
            },
            {
                "질문": "🏥 국민건강보험공단에서 실제로 건강검진 안내 문자를 보낼 때 포함되는 필수 정보는?",
                "보기": {
                    1: "홈페이지 주소와 대표전화(1577-1000)",
                    2: "개인 통장 계좌번호",
                    3: "QR코드와 약관 링크"
                },
                "정답": 1
            },
            {
                "질문": "⚖️ 정부기관은 문자나 전화로 직접 계좌 이체를 요구할 수 있다. (O/X)",
                "보기": {1: "O", 2: "X"},
                "정답": 2
            }
        ]
    },

    "가족/지인 사칭": {
        "설명": "가족이나 지인을 사칭해 긴급 상황을 연출하고 금전 송금을 유도하는 스미싱입니다.",
        "문제": [
            {
                "질문": "👩‍🦰 '엄마 나 휴대폰 고장났어. 급하게 돈 좀 보내줘'라는 문자를 받았다면?",
                "보기": {
                    1: "보낸 번호로 송금한다.",
                    2: "전화로 직접 자녀 여부를 확인한다.",
                    3: "카카오톡으로 확인한다."
                },
                "정답": 2
            },
            {
                "질문": "💰 자녀 사칭 스미싱이 가장 잘 통하는 이유는?",
                "보기": {
                    1: "시간이 늦은 밤에 자주 보내서",
                    2: "가족 간의 신뢰와 긴급성을 이용하기 때문",
                    3: "대부분 발신 번호가 010으로 시작하기 때문"
                },
                "정답": 2
            },
            {
                "질문": "💌 모바일 청첩장 스미싱의 주요 특징은?",
                "보기": {
                    1: "경조사와 관련된 친근한 내용으로 링크 클릭 유도",
                    2: "해외 배송 관련 안내로 위장",
                    3: "금융 기관 로그인 페이지로 연결"
                },
                "정답": 1
            },
        ]
    },

    "금융/대출 사칭": {
        "설명": "저금리 대출, 정부지원금, 대출 상환 유예 등으로 속이는 금융 피싱입니다.",
        "문제": [
            {
                "질문": "🏦 은행은 고객에게 문자로 대출 권유를 보낼 수 있다. (O/X)",
                "보기": {1: "O", 2: "X"},
                "정답": 2
            },
            {
                "질문": "💸 '정부 긴급대출 신청하세요. 오늘 마감입니다!' 문자의 대표적 특징은?",
                "보기": {
                    1: "시간 압박과 긴급성 유도",
                    2: "사실 확인이 가능한 정보 제공",
                    3: "공식기관의 문체 사용"
                },
                "정답": 1
            },
            {
                "질문": "📲 금융 스미싱의 가장 흔한 수법은?",
                "보기": {
                    1: "앱 설치 유도 후 악성코드 감염",
                    2: "공식 앱 업데이트 요청",
                    3: "상품권 지급 이벤트"
                },
                "정답": 1
            }
        ]
    },

    "보이스피싱 심화": {
        "설명": "심리전·설득·긴급 압박을 활용한 실전화 기반 피싱 유형입니다.",
        "문제": [
            {
                "질문": "📞 '당신의 계좌가 범죄에 이용되었습니다. 바로 이체해야 보호됩니다.' 어떤 심리를 노린 수법일까요?",
                "보기": {
                    1: "공포와 불안",
                    2: "보상 심리",
                    3: "관심 유도"
                },
                "정답": 1
            },
            {
                "질문": "👮‍♂️ 경찰관을 사칭하는 피싱 전화는 어떤 특징을 가질까요?",
                "보기": {
                    1: "정확한 직함과 공문서를 언급",
                    2: "통화 중 송금 또는 이체를 요구",
                    3: "경찰서 방문 요청"
                },
                "정답": 2
            },
            {
                "질문": "🤔 '피싱 피해 예방센터입니다. 고객님의 정보를 재확인하려 합니다.' 이 문장은 어떤 위험 신호일까요?",
                "보기": {
                    1: "공식 기관의 안내일 수 있다",
                    2: "개인정보 탈취 목적일 가능성이 높다",
                    3: "통신사 점검 과정일 수 있다"
                },
                "정답": 2
            }
        ]
    }
}


# ====================================
# 3️⃣ Gemini 피드백 (틀렸을 때)
# ====================================
def ai_feedback(question, user_choice):
    prompt = f"""
    너는 보이스피싱 예방 교사야.
    문제: {question}
    사용자의 선택: {user_choice}
    왜 위험한 선택인지 1~2문장으로 설명해줘.
    """
    try:
        resp = client.models.generate_content(model="gemini-2.5-flash", contents=prompt)
        return resp.text.strip()
    except:
        return "⚠️ AI 피드백 생성 오류 (API 키 또는 연결 문제)"

# ====================================
# 4️⃣ AI 칭찬 메시지 (레벨업 시)
# ====================================
def ai_praise(level):
    prompt = f"""
    사용자가 보이스피싱 예방 학습에서 레벨 {level}을 달성했습니다.
    격려의 한마디를 한국어로 1~2문장 써주세요.
    """
    try:
        resp = client.models.generate_content(model="gemini-2.5-flash", contents=prompt)
        return resp.text.strip()
    except:
        return f"🎉 축하합니다! SafeTalk 레벨 {level} 달성!"

# ====================================
# 5️⃣ 퀴즈 실행
# ====================================
def phishing_quiz(user_id):
    print("\n🧠 SafeTalk 보이스피싱 예방 퀴즈\n")
    for i, key in enumerate(scenarios.keys(), 1):
        print(f"{i}. {key} — {scenarios[key]['설명']}")
    try:
        choice = int(input("\n번호 입력: "))
        category = list(scenarios.keys())[choice - 1]
    except:
        print("잘못된 입력입니다.")
        return

    score, exp_gain = 0, 0
    print(f"\n🎯 [{category}] 시나리오 시작!\n")

    for q in scenarios[category]["문제"]:
        print(q["질문"])
        for k, v in q["보기"].items():
            print(f"{k}. {v}")
        ans = int(input("👉 선택 (번호): "))

        if ans == q["정답"]:
            print("✅ 정답입니다! +5점, +5 EXP\n")
            score += 5
            exp_gain += 5
        else:
            print("❌ 오답이에요. +1 EXP\n")
            fb = ai_feedback(q["질문"], q["보기"][ans])
            print("💬 AI 피드백:", fb, "\n")
            exp_gain += 1

    total = len(scenarios[category]["문제"]) * 5
    cursor.execute("INSERT INTO results (user_id, scenario, score, total) VALUES (?, ?, ?, ?)",
                   (user_id, category, score, total))
    conn.commit()

    # 업데이트
    cursor.execute("SELECT exp, level, total_score FROM users WHERE id=?", (user_id,))
    exp, level, total_score = cursor.fetchone()

    new_exp = exp + exp_gain
    new_score = total_score + score
    new_level = level + new_exp // 20  # 20 EXP마다 레벨업
    new_exp = new_exp % 20

    cursor.execute("UPDATE users SET exp=?, level=?, total_score=? WHERE id=?",
                   (new_exp, new_level, new_score, user_id))
    conn.commit()

    print(f"🏁 퀴즈 종료! 이번 점수: {score}/{total} | 누적 점수: {new_score} | EXP: {new_exp}/20")

    if new_level > level:
        praise = ai_praise(new_level)
        print(f"\n🌟 레벨업! 현재 레벨: {new_level}\n{praise}\n")

# ====================================
# 6️⃣ 통계 보기
# ====================================
def show_stats(user_id):
    df = pd.read_sql_query(
        f"SELECT scenario, AVG(CAST(score AS FLOAT)/total)*100 AS accuracy FROM results WHERE user_id={user_id} GROUP BY scenario",
        conn
    )
    user = cursor.execute("SELECT level, exp, total_score FROM users WHERE id=?", (user_id,)).fetchone()
    print(f"\n📊 현재 레벨: {user[0]} | EXP: {user[1]}/20 | 누적 점수: {user[2]}")
    print(df if not df.empty else "아직 기록이 없습니다.")

# ====================================
# 7️⃣ 실행
# ====================================
name = input("👤 사용자 이름을 입력하세요: ").strip()
cursor.execute("INSERT OR IGNORE INTO users (name) VALUES (?)", (name,))
conn.commit()
user_id = cursor.execute("SELECT id FROM users WHERE name=?", (name,)).fetchone()[0]

print(f"\n환영합니다, {name}님! SafeTalk 레벨 시스템이 활성화되었습니다 🎮")

while True:
    print("\n1️⃣ 퀴즈 시작\n2️⃣ 통계 보기\n3️⃣ 종료")
    sel = input("선택: ")
    if sel == "1":
        phishing_quiz(user_id)
    elif sel == "2":
        show_stats(user_id)
    elif sel == "3":
        print("👋 이용해주셔서 감사합니다!")
        break
    else:
        print("잘못된 입력입니다.")
```
